```python
Y = data['Rent'].values
X1 = data['Size'].values
X2 = data['BHK'].values
X3 = data['City'].replace({'Bangalore':1,'Chennai':2,'Delhi':3,'Hyderabad':4,'Kolkata':5,'Mumbai':6})
X4 = data['Bathroom'].values
X5 = data["Furnishing Status"].replace({"Unfurnished": 0, "Semi-Furnished": 1, "Furnished": 2}).values

print(X3)
print(X5)
```

```python
def min_max_scalar(input_list):
  min = np.min(input_list)
  max = np.max(input_list)
  dif = max-min
  return [(item-min)/dif for item in input_list]

mean_X1 = min_max_scalar(X1)
mean_X2 = min_max_scalar(X2)
mean_X3 = min_max_scalar(X3)
mean_X4 = min_max_scalar(X4)
mean_X5 = min_max_scalar(X5)
```

```python
theta_0 = 0
theta_1 = 0
theta_2 = 0
theta_3 = 0
theta_4 = 0
theta_5 = 0


m = len(X1)
epoch = 1000
learning_rate = 0.00000001
for i in range(epoch):
  temp_0 = 0
  temp_1 = 0
  temp_2 = 0
  temp_3 = 0
  temp_4 = 0
  temp_5 = 0
  
  for j in range(m):
    temp_0 += (theta_0 + theta_1*mean_X1[j] + theta_2*mean_X2[j] + theta_3*mean_X3[j] + theta_4*mean_X4[j] + theta_5*mean_X5[j] - Y[j])
    temp_1 += (theta_0 + theta_1*mean_X1[j] + theta_2*mean_X2[j] + theta_3*mean_X3[j] + theta_4*mean_X4[j] + theta_5*mean_X5[j] - Y[j]) * mean_X1[j]
    temp_2 += (theta_0 + theta_1*mean_X1[j] + theta_2*mean_X2[j] + theta_3*mean_X3[j] + theta_4*mean_X4[j] + theta_5*mean_X5[j] - Y[j]) * mean_X2[j]
    temp_3 += (theta_0 + theta_1*mean_X1[j] + theta_2*mean_X2[j] + theta_3*mean_X3[j] + theta_4*mean_X4[j] + theta_5*mean_X5[j] - Y[j]) * mean_X3[j]
    temp_4 += (theta_0 + theta_1*mean_X1[j] + theta_2*mean_X2[j] + theta_3*mean_X3[j] + theta_4*mean_X4[j] + theta_5*mean_X5[j] - Y[j]) * mean_X4[j]
    temp_5 += (theta_0 + theta_1*mean_X1[j] + theta_2*mean_X2[j] + theta_3*mean_X3[j] + theta_4*mean_X4[j] + theta_5*mean_X5[j] - Y[j]) * mean_X5[j]
    

  theta_0 = theta_0 - learning_rate*temp_0/m
  theta_1 = theta_1 - learning_rate*temp_1/m
  theta_2 = theta_2 - learning_rate*temp_2/m
  theta_3 = theta_3 - learning_rate*temp_3/m
  theta_4 = theta_4 - learning_rate*temp_4/m
  theta_5 = theta_5 - learning_rate*temp_5/m
  
print(theta_0,theta_1,theta_2,theta_3,theta_4,theta_5)
```


**Cost Calculation:**

```python
sum = 0
for j in range(m):
  err = theta_0 + theta_1*mean_X1[j] + theta_2*mean_X2[j] + theta_3*mean_X3[j] + theta_4*mean_X4[j] + theta_5*mean_X5[j]
  err = err** 2
  sum+=err


cost = sum/(2*m)

print("Cost :", cost)
```
